```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

## Dados dos algoritmos

```python
nomes = ["Força Bruta", "Guloso", "Dinâmico"]

entradas = [10, 100, 1000, 5000, 10000]
'''
algoritmo1 = [0.000334, 0.000044, 0.003519, 0.015023, 3.258980]
algoritmo2 = [0.000005, 0.000141, 0.010365, 0.094830, 0.343151]
algoritmo3 = [0.000068, 0.000494, 0.005352, 0.017810, 0.025452]
'''

df = pd.DataFrame(columns=["10", "100", "1000", "5000", "10000", "Algoritmo"])

df.loc[0] = [0.000334, 0.000044, 0.003519, 0.015023, 3.258980, "Força Bruta"]
df.loc[1] = [0.000005, 0.000141, 0.010365, 0.094830, 0.343151, "Guloso"]
df.loc[2] = [0.000068, 0.000494, 0.005352, 0.017810, 0.025452, "Dinâmico"]

df.head()
```

```python
# Transformando o DataFrame para um formato longo (melt)
df_melted = df.melt(id_vars=["Algoritmo"], var_name="Entrada", value_name="Tempo")

# Configurando o gráfico
plt.figure(figsize=(12, 8))
sns.set(style="whitegrid")

# Gráfico de barras
sns.barplot(x="Entrada", y="Tempo", hue="Algoritmo", data=df_melted)

# Personalizando o gráfico
plt.title('Comparação de Tempo de Execução dos Algoritmos')
plt.xlabel('Tamanho da Entrada')
plt.ylabel('Tempo de Execução (em segundos)')

# Adicionando os valores exatos no topo das barras
for index, row in df_melted.iterrows():
    plt.text(row["Entrada"], row["Tempo"], f'{row["Tempo"]:.6f}',
             horizontalalignment='center', size='small', color='black', weight='semibold')

plt.grid(True)
plt.yscale('log')  # Usando escala logarítmica no eixo y, se necessário
plt.show()
```

## Gerar Casos Aleatórios

```python
import random

def gerar_casos(n):
  itens = []

  for i in range(n):
    peso = random.randint(1, 100)
    valor = random.randint(1, 100)
    itens.append({"p": peso, "v": valor})

  return itens
```

```python
#usando as entradas

for entrada in entradas:
  itens = gerar_casos(entrada)
  capacidade = random.randint(200, 300)

  with open(f"{entrada}.txt", "w") as f:
    f.write(f"{len(itens)} {capacidade}\n")
    for item in itens:
      f.write(str(item["p"]) + " " + str(item["v"]) + "\n")
```
